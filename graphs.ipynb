```python
import os
import scipy
import pandas as pd
import matplotlib.pyplot as plot
import numpy as np

def pullData(file):
    f = pd.read_csv(file)
    f = f['RSSI']
    out = list(f)
    del out[0]
    out = list(map(int, out))
    return out

directory = 'ScannerOutputData'
rssi_files = []
for filename in os.listdir(directory):
    full_path = os.path.join(directory, filename)
    rssi_files.append(full_path)

real_rssi_values = []
for file in rssi_files:
    real_rssi_values.append(pullData(file))
```

```python
y1 = real_rssi_values[7:24]
x1 = [9.33, 3.55, 1.8, 1.03, .33, .03, 4.55, 6.54, 3.1, 5.0, 8.0, 8.5, 7.5, 6.0, 5.5, 2.0, 4.0]

for xe, ye in zip(x1, y1):
    plot.scatter([xe] * len(ye), ye, c='#146eff')
```
